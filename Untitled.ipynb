```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import tensorflow as tf
import glob
import cv2
from PIL import Image
from natsort import natsorted
from tqdm import tqdm_notebook
import utils
import os
```

```
img_list, mask_list = utils.load_data('data')
print("Number of training images: {}".format(len(img_list)))
print("Number of masking images: {}".format(len(mask_list)))
```

```
gen = utils.Datagen(img_list, mask_list, (1280, 1918, 3), 5, 1, shuffle=False) 
x = gen.generate_train_batch()
```

```
for _ in range(5):

    plt.figure(figsize=(16,8))

    batch_images, batch_masks = next(x)

    img_names = [os.path.basename(gen.img_list[ind]) for ind in gen.batch_rows]
    mask_names = [os.path.basename(gen.mask_list[ind]) for ind in gen.batch_rows]

    for i in range(gen.batch_size):

        plt.subplot(1, len(batch_images), i + 1)

        plt.title(img_names[i])

        plt.imshow(batch_images[i])

    plt.show()

    plt.figure(figsize=(16,8))

    for i in range(gen.batch_size):

        plt.subplot(1, len(batch_images), i + 1)

        plt.title(mask_names[i])

        plt.imshow(np.argmax(batch_masks[i], axis=2))

    plt.show()

    for i in range(gen.batch_size):

        plt.subplot(1, len(batch_images), i + 1)

        plt.imshow(cv2.bitwise_and(batch_images[i], batch_images[i], batch_masks[i]))

    plt.show()
```
